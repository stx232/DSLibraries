```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.fillna(df.median(), inplace=True)
```

```python
df.info()
```

```python
df = pd.get_dummies(df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

lr = LinearRegression()

x = df.drop('Price', axis=1)
y = df['Price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=5)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
pd.DataFrame({'price': y_test, 'predicted': y_pred})
```

```python
r2_score(y_test, y_pred)
```

```python
y_pred_train = lr.predict(x_train)
```

```python
r2_score(y_train, y_pred_train)
```

Модель не переобучена, но результат плохой.

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=0)
```

```python
model.fit(x_train, y_train.values)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
y_pred_train = model.predict(x_train)
r2_score(y_train, y_pred_train)
```

Метрика гораздо лучше, чем у линейной модели. Показатель на обучающей выборке очень высокий, но вроде для случайного леса это нормально. 

```python
from sklearn.model_selection import GridSearchCV

params = [{'min_samples_leaf': np.arange(1, 6),
           'max_features': [3, 4, 5, 'auto']}]
```

```python
model = GridSearchCV(estimator=RandomForestRegressor(random_state=0, n_jobs=-1),
                     param_grid=params, scoring='r2', cv=3)
```

model.fit(x_train, y_train)

```python
model.best_params_
```
